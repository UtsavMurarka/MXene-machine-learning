```python
print("test")
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
print("test")

#data=pd.read_csv("C:\\Users\\Utsav\\Desktop\\Mxene\\ML_exp", sep=',',header=0)

url = "https://raw.githubusercontent.com/UtsavMurarka/MXene-machine-learning/master/data_new.csv"

data=pd.read_csv(url, sep=',',header=0)
data
```

```python
mxene=data.values
mxene=np.array(mxene)
```

```python
#split the data

train, test = train_test_split(data, test_size=0.2)
train=np.array(train)
test=np.array(test)
```

```python
NegativeCount=0
PositiveCount=0
for i in range(len(train)):
    if train[i][30]==0 :
        NegativeCount=NegativeCount+1
    if train[i][30]==1:
        PositiveCount=PositiveCount+1
print(NegativeCount)
print(PositiveCount)
print(PositiveCount+NegativeCount)
```

```python
positives=np.zeros((PositiveCount,31))
negatives=np.zeros((NegativeCount,31))
    
j=0
k=0
    
for i in range(len(train)):

    if(train[i,30] == 1):
        positives[j,:] = train[i,:]
        j=j+1
    if(train[i,30] == 0):
        negatives[k,:] = train[i,:]
        k=k+1

print(np.shape(train))
print(np.shape(test))
```

```python
#Visualization
```

```python
def plot(ind):
  bins=20
  clr=['r', 'b']
  lbl=['1', '0']
  x=[positives[:,ind], negatives[:,ind]]
  plt.hist(x, bins, density=True, color=clr, label=lbl)
  plt.legend(loc='upper right')
  plt.title('Density')
  plt.show()
  #
  # bins = np.linspace(-1, 250, 50)
  # plt.hist(positives[:,ind], bins, alpha=0.5, label='1', color='r', density=True)
  # plt.hist(negatives[:,ind], bins, alpha=0.5, label='0', color='b', density=True)
  # plt.legend(loc='upper right')
  # plt.title('Density overlap')
  # plt.show()
  #
  # bins = np.linspace(-1, 250, 50)
  # plt.hist(positives[:,ind], bins, alpha=0.5, label='1', color='r')
  # plt.hist(negatives[:,ind], bins, alpha=0.5, label='0', color='b')
  # plt.legend(loc='upper right')
  # plt.title('without Density overlap')
  # plt.show()
  #
  # bins=100
  # plt.hist(positives[:,ind], bins)
  # plt.title('only postives (Not density)')
  # plt.show()
  #
  # bins=100
  # plt.hist(negatives[:,ind],bins)
  # plt.title('only negatives (Not density)')
  # plt.show()
  #
  print("Mean for positive data = ", np.mean(positives[:,ind]))
  print("Mean for negative data = ", np.mean(negatives[:,ind]))
```

```python
for i in range(31):
  plot(i)
```

```python
y_train=np.zeros(len(train))
y_test=np.zeros(len(test))
```

```python
for i in range(len(test)):
    y_test[i]=test[i][30]
    test[i][30]=1

for i in range(len(train)):
    y_train[i]=train[i][30]
    train[i][30]=1

train = np.delete(train, 30, 1)
test = np.delete(test, 30, 1)

train = np.delete(train, 29, 1)
test = np.delete(test, 29, 1)

print(np.shape(train))
print(np.shape(test))


positives = sum(y_train)
negatives = len(y_train) - positives

print("test")
print("positives = %d \n negatives = %d" % (positives, negatives))

positives = sum(y_test)
negatives = len(y_test) - positives

print("train")
print("positives = %d \n negatives = %d" % (positives, negatives))

print(train)
```

```python
from sklearn.metrics import accuracy_score
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(train,y_train)
y_pred = clf.predict(test)
```

```python
print("Accuracy = " + str(100*accuracy_score(y_test, y_pred)) + "%")
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
naiveBayes = gnb.fit(train, y_train)
y_pred = naiveBayes.predict(test)
```

```python
print("Accuracy = " + str(100*accuracy_score(y_test, y_pred)) + "%")
```
