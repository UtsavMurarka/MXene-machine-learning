```
import urllib3
import numpy as np
import pandas as pd

import requests

def exists(path):
    r = requests.head(path, verify=False)
    return r.status_code == requests.codes.ok

http = urllib3.PoolManager()

for num in range(1000,100000):
  url = "http://anant.mrc.iisc.ac.in/mxenedb/%5Edetails/download_poscar/(" + str(num)+ ")"
  if (exists(url)):
      r = http.request('GET', url, preload_content=False)
      name = "poscar.txt"
      path = "./"+name
      with open(path, 'wb') as out:
          data = r.read()
          out.write(data)
      r.release_conn()
  else:
      print (str(num) + " "+" not found")
```
